```python
import yaml

with open("gd.yml") as f1, open("control.yml") as f2:
    env1 = yaml.safe_load(f1)
    env2 = yaml.safe_load(f2)

def separate_conda_and_pip(deps):
    conda = []
    pip = []
    for dep in deps:
        if isinstance(dep, dict) and "pip" in dep:
            pip += dep["pip"]
        elif isinstance(dep, str):
            conda.append(dep)
    return conda, pip

conda1, pip1 = separate_conda_and_pip(env1["dependencies"])
conda2, pip2 = separate_conda_and_pip(env2["dependencies"])

merged_conda = sorted(set(conda1 + conda2))
merged_pip = sorted(set(pip1 + pip2))

merged_deps = merged_conda
if merged_pip:
    merged_deps.append({"pip": merged_pip})

merged_channels = sorted(set(env1.get("channels", []) + env2.get("channels", [])))

merged_env = {
    "name": "merged_env",
    "channels": merged_channels,
    "dependencies": merged_deps
}

with open("merged_env.yml", "w") as f:
    yaml.dump(merged_env, f, sort_keys=False)
```
